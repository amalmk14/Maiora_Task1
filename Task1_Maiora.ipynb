```python
import pandas as pd 
import sqlite3
import json
```

```python
order_region_a = pd.read_csv("order_region_a.csv")
order_region_b = pd.read_csv("order_region_b.csv")
```

```python
order_region_a
```

```python
order_region_b
```

```python
order_region_a['region'] = 'A'
order_region_b['region'] = 'B'
```

```python
data1 = pd.concat([order_region_a, order_region_b], ignore_index=True)
data1
```

```python
# Total Sales
```

```python
data1['total_sales'] = data1['QuantityOrdered'] * data1['ItemPrice']
data1.head()
```

```python
data1['QuantityOrdered'] = pd.to_numeric(data1['QuantityOrdered'], errors='coerce').fillna(0)
data1['ItemPrice'] = pd.to_numeric(data1['ItemPrice'], errors='coerce').fillna(0)

def extract_discount(value):
    try:
        return float(json.loads(value)['Amount']) if isinstance(value, str) and value.startswith('{') else float(value)
    except (ValueError, TypeError, KeyError):
        return 0

data1['PromotionDiscount'] = data1['PromotionDiscount'].apply(extract_discount)
data1['PromotionDiscount']
```

```python
# Net Sale
```

```python
data1['net_sale'] = data1['total_sales'] - data1['PromotionDiscount']
```

```python
data1
```

```python
# Remove Duplicate in Order ID
```

```python
data1.drop_duplicates(subset=['OrderId'], keep='first', inplace=True)
data1
```

```python
data1 = data1[data1['net_sale'] > 0]
data1
```

```python
conn = sqlite3.connect("sales_data.db")
cursor = conn.cursor()
```

```python
cursor.execute('''
    CREATE TABLE IF NOT EXISTS sales_data (
        OrderId INTEGER PRIMARY KEY,
        OrderItemId INTEGER,
        QuantityOrdered INTEGER,
        ItemPrice REAL,
        PromotionDiscount REAL,
        total_sales REAL,
        net_sale REAL,
        region TEXT
    )
''')
```

```python
data1.to_sql("sales_data", conn, if_exists='replace', index=False)
```

```python
queries = {
    "Total Records": "SELECT COUNT(*) FROM sales_data",
    "Total Sales by Region": "SELECT region, SUM(total_sales) FROM sales_data GROUP BY region",
    "Average Sales Per Transaction": "SELECT AVG(total_sales) FROM sales_data",
    "Check Duplicate OrderId": "SELECT OrderId, COUNT(*) FROM sales_data GROUP BY OrderId HAVING COUNT(*) > 1"
}
```

```python
for desc, query in queries.items():
    cursor.execute(query)
    print(f"{desc}: {cursor.fetchall()}")
```

```python
conn.close()
```
